```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import math
```

```python
ab_train = pd.read_csv('train-abnormal.csv')
ab_val = pd.read_csv('valid-abnormal.csv')
acl_train = pd.read_csv('train-acl.csv')
acl_val = pd.read_csv('valid-acl.csv')
men_train = pd.read_csv('train-meniscus.csv')
men_val = pd.read_csv('valid-meniscus.csv')
ab_cat = pd.concat((ab_train, ab_val), axis = 0)
acl_cat = pd.concat((acl_train, acl_val), axis=0)
men_cat = pd.concat((men_train, men_val), axis=0)
print("length of train set : ", len(ab_train))
print("length of test set : ", len(ab_val))
print("total length of train + test : ", len(ab_cat))
```

```python
df = pd.DataFrame({'abnormal':ab_cat['label'], 'acl':acl_cat['label'], 'meniscus':men_cat['label']})
df
```

```python
axial_num_slides = []
coronal_num_slides = []
sagittal_num_slides = []

for i in range(0, 1130, 1):
    if i == 0:
        i = 1
    num_zeros = 3 - int(math.log10(i))
    zeros = ["0" for _ in range(num_zeros)]
    zeros = ''.join(zeros)    
    ex = np.load(f'./train/axial/{zeros}{i}.npy')
    axial_num_slides.append(ex.shape[0])

for i in range(0, 1130, 1):
    if i == 0:
        i = 1
    num_zeros = 3 - int(math.log10(i))
    zeros = ["0" for _ in range(num_zeros)]
    zeros = ''.join(zeros)    
    ex = np.load(f'./train/coronal/{zeros}{i}.npy')
    coronal_num_slides.append(ex.shape[0])

for i in range(0, 1130, 1):
    if i == 0:
        i = 1
    num_zeros = 3 - int(math.log10(i))
    zeros = ["0" for _ in range(num_zeros)]
    zeros = ''.join(zeros)    
    ex = np.load(f'./train/sagittal/{zeros}{i}.npy')
    sagittal_num_slides.append(ex.shape[0])


for i in range(1130, 1250, 1):
    
    num_zeros = 3 - int(math.log10(i))
    zeros = ["0" for _ in range(num_zeros)]
    zeros = ''.join(zeros)    
    ex = np.load(f'./valid/axial/{zeros}{i}.npy')
    axial_num_slides.append(ex.shape[0])

for i in range(1130, 1250, 1):
    
    num_zeros = 3 - int(math.log10(i))
    zeros = ["0" for _ in range(num_zeros)]
    zeros = ''.join(zeros)    
    ex = np.load(f'./valid/coronal/{zeros}{i}.npy')
    coronal_num_slides.append(ex.shape[0])

for i in range(1130, 1250, 1):
    
    num_zeros = 3 - int(math.log10(i))
    zeros = ["0" for _ in range(num_zeros)]
    zeros = ''.join(zeros)    
    ex = np.load(f'./valid/sagittal/{zeros}{i}.npy')
    sagittal_num_slides.append(ex.shape[0])

print(len(axial_num_slides))
print(len(coronal_num_slides))
print(len(sagittal_num_slides))
df_slides = pd.DataFrame({'axial':axial_num_slides, 'coronal':coronal_num_slides, 'sagittal':sagittal_num_slides})
df_slides.tail(10)
```
